```python
import os
from dotenv import load_dotenv

load_dotenv()
MONGODB_ATLAS_CLUSTER_URI = os.getenv('MONGODB_URI')
```

```python
from pymongo.mongo_client import MongoClient
from pymongo.server_api import ServerApi


# Create a new client and connect to the server
client = MongoClient(MONGODB_ATLAS_CLUSTER_URI, server_api=ServerApi('1'))

# Send a ping to confirm a successful connection
try:
    client.admin.command('ping')
    print("Pinged your deployment. You successfully connected to MongoDB!")
except Exception as e:
    print(e)
```

```python

db = client["poc-nero"]
noticias_collection = db["noticias"]

# Consulta todos os documentos na coleção
noticias = noticias_collection.find()

# Exibe os documentos
for noticia in noticias:
    print(noticia)
    break
```
